```python
import re
def clean_date(v):
    date = re.search('\d{4}', v)
    if date:
        return int(date.group())
    return 'missing'
    
df['date_norm'] = df.date.apply(clean_date)

df.date_norm.unique()
```

```python
df = df[df.date_norm != 'missing']
```

```python
df.plot(x='acquisition_date', y='date_norm', kind='scatter', figsize=(50, 40))
```
